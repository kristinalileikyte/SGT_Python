## 1) Load a dataset using Pandas and perform data cleaning operations such as removing null values, removing duplicates,number format etc. ##

```python
data = pandas.read_excel('https://unstats.un.org/unsd/envstats/Questionnaires/2019/Tables/Precipitation.xlsx','data')
```

```python
data
```

```python
data.drop(data.iloc[:, 30:], inplace=True, axis=1)
data
```

```python
data.replace('...',np.NaN, inplace=True)
years = data.columns[2:30]
data.dropna(subset=years,how='all',inplace = True)
data
```

```python
data.drop_duplicates()
```

```python
pandas.options.display.float_format = '{:,.1f}'.format
data
```

## 6) Use Pandas to merge two datasets together based on a common key. ##

```python
merged_table = pandas.merge(data, new_data, on='Country', how='outer')
```
